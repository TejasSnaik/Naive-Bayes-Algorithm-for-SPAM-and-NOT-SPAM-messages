```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("SPAM text message 20170820 - Data.csv")
```

```python
df
```

```python
df11=pd.DataFrame(np.random.randn(4,2), index=list('0123'),columns=[1,2])
df11
```

```python
X = df.Category
```

```python
y= df.Message
```

```python
Message=pd.DataFrame(y,columns=['Message'])
Message.isnull().sum()
```

```python
scaler =  LabelEncoder()
```

```python
label_encoded_data=scaler.fit_transform(X)
```

# 

```python
D=label_encoded_data.reshape(-1,1)
Category=pd.DataFrame(D,columns =['Category'])
Category
```

```python
email=pd.concat([Category,Message],axis=1)
```

```python
email.isnull().sum()
```

```python

email.sort_index(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer(stop_words='english')
```

```python
features=vectorizer.fit_transform(email.Message)
```

```python
features.shape
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x,train_y,test_y=train_test_split(features,email.Category,test_size=0.3,random_state=88)
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
test_x.shape
```

```python
test_y.shape
```

```python
test_x
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier=MultinomialNB()
```

```python
classifier.fit(train_x,train_y)
```

```python
nrr_correct=(test_y==classifier.predict(test_x)).sum()
```

```python
test_y
```

```python
classifier.predict(test_x)
```

```python
nrr_correct
```

```python
nrr_nocorrect=test_y.size-nrr_correct
```

```python
nrr_nocorrect
```

```python
fraction_correct=nrr_correct/(nrr_correct+nrr_nocorrect)
```

```python
print('The tested accuracy of the model is',format(fraction_correct,".2%"))
```

```python
fraction_nocorrect=nrr_nocorrect/(nrr_correct+nrr_nocorrect)
```

```python
print(format(fraction_nocorrect,".2%"))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
recall_score(test_y,classifier.predict(test_x))
```

```python
precision_score(test_y, classifier.predict(test_x))
```

```python
f1_score(test_y, classifier.predict(test_x))
```
